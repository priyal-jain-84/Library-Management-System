```python
import random
dict_of_books={'101':['Invisible Man','Ralph Ellison',581,10,'9788122903928',1952],
               '102':['Beloved','Toni Morrison',324,15,'9781400033416',1987], 
               '103':['A Passage to India','E. M. Forster',368,12,'9780156711425',1924],
               '104':['The Great Gatsby','F. Scott Fitzgerald',180,16,'9780743273565',1925],
               '105':['Don Quixote','Miguel de Cervantes',462,11,'9780060188702',1605]}

dict_of_borrowers_id={}
borrowing_history={}

class LMS:
    def menu(self):
        a = input("Do you want to login or check the books available in library.... \n Please use 'Login' or 'Check Books'")
        if(a=='Login'):
            b = input('Are you a Admin or Borrower please reply with Admin or Borrower....')
            if (b=='Admin'):
                print('Please use Admin() class to login...')
                

            elif (b=='Borrower'):
                print('Please use Borrower() class to login....')
                

            else:
                print("Please use correct Stakeholder its either 'Admin' or 'Borrower'")

        elif(a=='Check Books'):
            for key,value in dict_of_books.items():
                print(key,value)
            
        else:
            print('Please choose correct option...')
            
```

```python
class Admin:
    dict_of_admin={}
    dict_of_login_admin={}

    def  __init__(self):
        print("Welcome Admin.... Please login or signup...")


    def sign_up(self):     
        self.name=input("Please enter your full name.... ")
        self.email=input("Please enter your email address.... ")
        self.password=input("Please enter your password.... ")
        
        if self.name not in Admin.dict_of_login_admin:
            self.dict_of_admin[self.name]=[self.email,self.password]
            print('Account created sucessfully')

        else:
            print('Name already exist..')
            
            
    def login(self,name):
        self.name=name
        email=input("Please enter your email address.... ")
        password=input("Please enter your password.... ")
                
        if self.name in Admin.dict_of_admin:
            if(self.email==email and self.password==password):
                Admin.dict_of_login_admin[self.name]=[self.email,self.password]
                print("Login successful")
            else:
                print("Incorrect login email and password")
        else:
            print("You don't have any account ,please sign-up")
            
    def logout(self,adname):
        self.adname=adname
        if self.adname in Admin.dict_of_login_admin:
            print("you are logged out")
            Admin.dict_of_login_admin.pop(self.name)
        else:
            print("you are not logged in")
            
            
    def add_borrower(self):
        self.admin_name=input("Please enter your name Admin...")
        if self.admin_name in Admin.dict_of_login_admin:
            self.full_name=input ('Enter borrowers name...')
            self.dob=input ('Enter borrowers date of birth...')
            self.contact_no=input ('Enter borrowers contact no...')
            self.email=input ('Enter borrowers Email Address...')
            self.password=input ('Enter borrowers password...')
            
            if self.full_name not in dict_of_borrowers_id:
                dict_of_borrowers_id[self.full_name]=[self.email,self.password,self.dob,self.contact_no]
                print('Borrower added succesfully...')
                
                
            else:
                print('Name already exist..')
            
        else:
            print('Please first login as an Admin...')
            
            
    def search_book(self):
        self.book_id= input('Enter the id of the book which you want to search....')
        
        if self.book_id in dict_of_books:
            print('Name of Book: ',dict_of_books[self.book_id][0])
            print('Author of the Book: ',dict_of_books[self.book_id][1])
            print('No. of pages: ',dict_of_books[self.book_id][2])
            print('No. of Books in the Library: ',dict_of_books[self.book_id][3])
            print('ISBN no. of Book: ',dict_of_books[self.book_id][4])
            print('This book was published in the year: ',dict_of_books[self.book_id][5])
            
        else:
            print('Please write the correct Book ID, or you can check it in menu.....')
            
            
    def add_book(self):
        self.admin_name=input("Please enter your name Admin...")
        if self.admin_name in Admin.dict_of_login_admin:
            self.book_name=input('Enter name of book...')
            self.author_name=input('Enter name of Author of the book...')
            self.no_of_pages=input('Enter no. of pages in the book...')
            self.book_in_lab=input('Enter no. of books in the library...')
            self.isbn=input('Enter ISBN no. of book...')
            self.publish_year=input('Enter the year in which the book was published...')
            self.book_id=str(random.randint(0,5000))
            dict_of_books[self.book_id]=[self.book_name,self.author_name,self.no_of_pages,self.book_in_lab,self.isbn,self.publish_year]
            print('Book added succesfully with book ID: ',self.book_id)
            print(dict_of_books)
        else:
            print('Please first login as an Admin...')
            
    def delete_book(self):
        admin_name=input("Please enter your name Admin...")
        if admin_name in Admin.dict_of_login_admin:
            self.id_of_book=input('Enter book id which you want to delete...')
            if self.id_of_book in dict_of_books:
                dict_of_books.pop(self.id_of_book)
                print('Book deleted successfully....')
            
            else:
                print('Please choose correct book id...')

        else:
            print('Please first login as an Admin...')
        
        
    def issue_book(self):
        self.admin_name=input("Please enter your name Admin...")
        if self.admin_name in Admin.dict_of_login_admin:
            self.borrower_name=input('Please enter full name of borrower..')
            self.borrower_email=input('Please enter email of borrower..')
            self.status="Issue"
            self.fine=0
            if self.borrower_name in dict_of_borrowers_id:            
                if self.borrower_email == dict_of_borrowers_id[self.borrower_name][0]:
                    self.book_id=input('Enter the book id which you want to issue...')
                    if self.book_id in dict_of_books:
                        if (int(dict_of_books[self.book_id][3]))>0:
                            borrowing_history[self.borrower_email]=[self.book_id,dict_of_books[self.book_id][0],self.status,self.fine]
                            print('Book issued...')
                            dict_of_books[self.book_id][3]-=1
                            print(dict_of_books)
                            print(borrowing_history)

                        else:
                            print('This book is currently out of stock...')

                    else:
                        print('Please select other book id... This book id is wrong..')


                else:
                    print('Incorrect email..')
                    
            else:
                print('Please register yourself as a borrower first...')

        else:
            print('Please login as an admin..')
            
    def deposit_book(self):
        self.admin_name=input("Please enter your name Admin...")
        if self.admin_name in Admin.dict_of_login_admin:
            self.borrower_email=input('Please enter email of borrower..')
            if self.borrower_email in borrowing_history:
                self.book_id=input('Enter the ID of the book which you want ot return....')
                if self.book_id == (borrowing_history[self.borrower_email][0]):
                    if borrowing_history[self.borrower_email][2]=='Issue':
                        self.days=int(input('Enter the no. of days since thi book was with you...'))
                        borrowing_history[self.borrower_email][2]='Deposit'
                        if self.days >15:
                            borrowing_history[self.borrower_email][3]=100
                            print('You have taken this book for more than 15 days hence you have to pay a penalty of Rs100.')
                            print('Book deposited....')
                            dict_of_books[self.book_id][3]+=1
                        else:
                            borrowing_history[self.borrower_email][3]=0
                            print('Thanks for returning the book on time...')
                            print('Book deposited....')
                            dict_of_books[self.book_id][3]+=1
                            
                            
                    else:
                        print('you already have deposited this book...')
                        
                else:
                    print('We have not issued this book to you...')
                    
                    
            else:
                print('You have not issued any book yet....')
                
                
        else:
            print('Please login as an admin..')
                            
                    
                    
```
